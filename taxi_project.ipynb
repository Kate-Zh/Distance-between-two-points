```python
def haversine(latitude1, longitude1, latitude2, longitude2, radius = 6371):
    # перевод градусов в радианы
    latitude1, longitude1, latitude2, longitude2 = [np.deg2rad(x) for x in (latitude1, longitude1, latitude2, longitude2)]
    # расчет расстояния между двумя точками Земли
    distance = radius * 2 * np.arcsin(np.sqrt((np.sin((latitude2-latitude1)/2)**2)+np.cos(latitude1)*np.cos(latitude2)*(np.sin((longitude2 - longitude1)/2)**2)))
    return distance
```

```python
distances = []
for i, row in nyc.iterrows():
    distance = haversine(row.pickup_latitude, row.pickup_longitude, row.dropoff_latitude, row.dropoff_longitude)
    distances.append(distance)
```
